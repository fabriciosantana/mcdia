## Carregamento de Dados de Contratações via API do PNCP

Este notebook demonstra como carregar dados de contratações públicas utilizando a API do PNCP (Portal Nacional de Contratações Públicas), especificamente o endpoint `/contratacoes`.

Para mais detalhes sobre a API, consulte o Swagger UI em: [https://dadosabertos.compras.gov.br/swagger-ui/index.html#/07%20-%20CONTRATA%C3%87%C3%95ES/consultarContratacoes_PNCP_14133](https://dadosabertos.compras.gov.br/swagger-ui/index.html#/07%20-%20CONTRATA%C3%87%C3%95ES/consultarContratacoes_PNCP_14133)

### 1. Importar Bibliotecas Necessárias

Vamos importar as bibliotecas `requests` para fazer as requisições HTTP e `pandas` para manipular os dados retornados em formato de DataFrame.

```python
import requests
import pandas as pd
```

### 2. Definir o Endpoint da API

O endpoint da API para consultar as contratações será definido. É importante garantir que a URL base esteja correta.

```python
base_url = "https://dadosabertos.compras.gov.br/modulo-contratacoes"
endpoint = "/1_consultarContratacoes_PNCP_14133"
api_url = f"{base_url}{endpoint}"
# Parâmetros de exemplo (ajuste conforme necessário para filtrar os dados)
# params = {
#     "offset": 0,
#     "limit": 10
# }

print(f"API URL: {api_url}")
print("https://dadosabertos.compras.gov.br/modulo-contratacoes/1_consultarContratacoes_PNCP_14133")
```

### 3. Fazer a Requisição à API

Utilizaremos a biblioteca `requests` para fazer uma requisição GET ao endpoint definido. Por padrão, a API do PNCP geralmente não exige autenticação para consultas públicas, mas pode-se adicionar parâmetros de consulta se necessário.

```python
try:
    params = {"pagina":1,
              "tamanhoPagina":10,
              "dataPublicacaoPncpInicial":"2026-06-01",
              "dataPublicacaoPncpFinal":"2026-07-05",
              "codigoModalidade":5  #pregao
    }


    response = requests.get(api_url, params)
    response.raise_for_status() # Levanta um erro para status codes HTTP ruins (4xx ou 5xx)
    data = response.json()
    print("Dados carregados com sucesso da API.")
except requests.exceptions.RequestException as e:
    print(f"Erro ao fazer a requisição à API: {e}")
    data = None
```

### 4. Processar e Exibir os Dados

Após receber a resposta da API (que é um JSON), vamos convertê-la em um DataFrame do pandas para facilitar a visualização e análise. O JSON retornado pela API do PNCP geralmente contém uma chave como 'data' ou 'result' que contém a lista de registros.

```python
if data:
    # A estrutura do JSON pode variar, geralmente os dados vêm dentro de uma chave
    # Verificando as chaves mais comuns, ou ajustando conforme o 'schema' do Swagger
    if 'resultado' in data:
        contratacoes_list = data['resultado']
    else:
        contratacoes_list = data # Assumindo que o JSON é diretamente uma lista

    if isinstance(contratacoes_list, list) and len(contratacoes_list) > 0:
        df = pd.DataFrame(contratacoes_list)
        print(f"DataFrame criado com {len(df)} linhas e {len(df.columns)} colunas.")
        print("Primeiras 5 linhas do DataFrame:")
        display(df.head())
    else:
        print("Nenhum dado de contratação encontrado ou formato inesperado.")
else:
    print("Não foi possível processar os dados porque a requisição à API falhou.")
```

### 5. Obter Links de Arquivos da Primeira Contratação

Agora que temos os dados das contratações, vamos usar o novo Swagger (https://pncp.gov.br/api/pncp/swagger-ui/index.html?configUrl=/pncp-api/v3/api-docs/swagger-config#/) para obter os links dos arquivos relacionados à primeira contratação do nosso DataFrame.

Para isso, precisaremos extrair o `orgaoEntidadeCnpj`, `anoCompraPncp` e o mês da `dataPublicacaoPncpInicial` (que foi usado como filtro na requisição anterior) da primeira contratação.

```python
if df is not None and not df.empty:
    first_contracting = df.iloc[0]
    cnpj_orgao = first_contracting['orgaoEntidadeCnpj']
    ano_compra = first_contracting['anoCompraPncp']
    sequencial_compra = first_contracting['sequencialCompraPncp']

    # Usando o mês do filtro da requisição anterior para construir a URL
    # ou pode-se tentar extrair do 'dataPublicacaoPncpInicial' se estiver disponível no df
    # No entanto, a API requer um mês específico, não um intervalo.
    # Considerando o filtro anterior: dataPublicacaoPncpInicial='2026-06-01', usaremos '6'
    print(f"Detalhes da primeira contratação para buscar arquivos:")
    print(f"  CNPJ do Órgão: {cnpj_orgao}")
    print(f"  Ano da Compra: {ano_compra}")
    print(f"  Sequencial: {sequencial_compra}")
else:
    print("Não há dados de contratação para extrair informações.")
    cnpj_orgao = None
    ano_compra = None
```

### 5.1. Construir a URL do Endpoint de Arquivos

Com os dados extraídos, montamos a URL para o endpoint de arquivos, que segue o padrão: `https://pncp.gov.br/api/pncp/v1/orgaos/{cnpj_orgao}/compras/{ano_compra}/{sequencial_compra}/arquivos`.

```python
if cnpj_orgao and ano_compra and sequencial_compra:
    files_api_url = f"https://pncp.gov.br/api/pncp/v1/orgaos/{cnpj_orgao}/compras/{ano_compra}/{sequencial_compra}/arquivos"
    print(f"URL para buscar arquivos: {files_api_url}")
else:
    print("Não foi possível construir a URL para buscar arquivos devido à falta de dados.")
    files_api_url = None
```

### 5.2. Fazer a Requisição para Obter os Links dos Arquivos

Realizamos a requisição GET para a URL dos arquivos e processamos a resposta.

```python
file_links = []
if files_api_url:
    try:
        # Os parâmetros devem estar incluídos diretamente na files_api_url, não como um dicionário 'params' separado.
        response_files = requests.get(files_api_url)
        response_files.raise_for_status()
        files_data = response_files.json()

        if files_data and isinstance(files_data, list):
            for file_info in files_data:
                if 'url' in file_info:
                    file_links.append(file_info['url'])
            print(f"Links de arquivos encontrados: {len(file_links)}")
            if file_links:
                print("Primeiros 5 links de arquivos:")
                for link in file_links[:5]:
                    print(link)
            else:
                print("Nenhum link de arquivo encontrado para esta contratação.")
        else:
            print("Resposta da API de arquivos inesperada ou vazia.")

    except requests.exceptions.RequestException as e:
        print(f"Erro ao buscar links de arquivos: {e}")
else:
    print("Não foi possível buscar links de arquivos porque a URL não foi construída.")
```
