## Carregamento de Dados de Contratações via API do PNCP

Este notebook demonstra como carregar dados de contratações públicas utilizando a API do PNCP (Portal Nacional de Contratações Públicas), especificamente o endpoint `/contratacoes`.

Para mais detalhes sobre a API, consulte o Swagger UI em: [https://dadosabertos.compras.gov.br/swagger-ui/index.html#/07%20-%20CONTRATA%C3%87%C3%95ES/consultarContratacoes_PNCP_14133](https://dadosabertos.compras.gov.br/swagger-ui/index.html#/07%20-%20CONTRATA%C3%87%C3%95ES/consultarContratacoes_PNCP_14133)

### 1. Importar Bibliotecas Necessárias

Vamos importar as bibliotecas `requests` para fazer as requisições HTTP e `pandas` para manipular os dados retornados em formato de DataFrame.

```python
import requests
import pandas as pd
import json
```

### 2. Definir o Endpoint da API

O endpoint da API para consultar as contratações será definido. É importante garantir que a URL base esteja correta.

```python
base_url = "https://dadosabertos.compras.gov.br/modulo-contratacoes"
endpoint = "/1_consultarContratacoes_PNCP_14133"
api_url = f"{base_url}{endpoint}"
# Parâmetros de exemplo (ajuste conforme necessário para filtrar os dados)
params = {
    "pagina":1,
    "tamanho":10,
    "dataPublicacaoPncpInicial":"2025-01-01",
    "dataPublicacaoPncpFinal":"2025-12-31",
    "orgaoEntidadeCnpj":"00530279000115",
    "codigoModalidade":5  #pregao
}
#     "offset": 0,
#     "limit": 10
# }

print(f"API URL: {api_url}")
print("https://dadosabertos.compras.gov.br/modulo-contratacoes/1_consultarContratacoes_PNCP_14133")

print("Parâmetros de consulta:")
for key, value in params.items():
    print(f"{key}: {value}")
```

### 3. Fazer a Requisição à API

Utilizaremos a biblioteca `requests` para fazer uma requisição GET ao endpoint definido. Por padrão, a API do PNCP geralmente não exige autenticação para consultas públicas, mas pode-se adicionar parâmetros de consulta se necessário.

```python
try:
    response = requests.get(api_url, params)
    response.raise_for_status() # Levanta um erro para status codes HTTP ruins (4xx ou 5xx)
    data = response.json()
    print("Dados carregados com sucesso da API.")
except requests.exceptions.RequestException as e:
    print(f"Erro ao fazer a requisição à API: {e}")
    data = None
```

### 4. Processar e Exibir os Dados

Após receber a resposta da API (que é um JSON), vamos convertê-la em um DataFrame do pandas para facilitar a visualização e análise. O JSON retornado pela API do PNCP geralmente contém uma chave como 'data' ou 'result' que contém a lista de registros.

```python
if data:
    # A estrutura do JSON pode variar, geralmente os dados vêm dentro de uma chave
    # Verificando as chaves mais comuns, ou ajustando conforme o 'schema' do Swagger
    if 'resultado' in data:
        contratacoes_list = data['resultado']
    else:
        contratacoes_list = data # Assumindo que o JSON é diretamente uma lista

    if isinstance(contratacoes_list, list) and len(contratacoes_list) > 0:
        df = pd.DataFrame(contratacoes_list)
        print(f"DataFrame criado com {len(df)} linhas e {len(df.columns)} colunas.")
        print("Primeiras 5 linhas do DataFrame:")
        display(df.head())
    else:
        print("Nenhum dado de contratação encontrado ou formato inesperado.")
else:
    print("Não foi possível processar os dados porque a requisição à API falhou.")
```

### 5. Obter Links de Arquivos da Primeira Contratação

Agora que temos os dados das contratações, vamos usar o novo Swagger (https://pncp.gov.br/api/pncp/swagger-ui/index.html?configUrl=/pncp-api/v3/api-docs/swagger-config#/) para obter os links dos arquivos relacionados à primeira contratação do nosso DataFrame.

Para isso, precisaremos extrair o `orgaoEntidadeCnpj`, `anoCompraPncp` e o mês da `dataPublicacaoPncpInicial` (que foi usado como filtro na requisição anterior) da primeira contratação.

### 5.2. Fazer a Requisição para Obter os Links dos Arquivos

Realizamos a requisição GET para a URL dos arquivos e processamos a resposta.

```python
file_links = []
if files_api_url:
    try:
        # Os parâmetros devem estar incluídos diretamente na files_api_url, não como um dicionário 'params' separado.
        response_files = requests.get(files_api_url)
        response_files.raise_for_status()
        files_data = response_files.json()

        if files_data and isinstance(files_data, list):
            for file_info in files_data:
                if 'url' in file_info:
                    file_links.append(file_info['url'])
            print(f"Links de arquivos encontrados: {len(file_links)}")
            if file_links:
                print("Primeiros 5 links de arquivos:")
                for link in file_links[:5]:
                    print(link)
            else:
                print("Nenhum link de arquivo encontrado para esta contratação.")
        else:
            print("Resposta da API de arquivos inesperada ou vazia.")

    except requests.exceptions.RequestException as e:
        print(f"Erro ao buscar links de arquivos: {e}")
else:
    print("Não foi possível buscar links de arquivos porque a URL não foi construída.")
```

```python
import json
import requests
import pandas as pd

registros_log = []

if df is not None and not df.empty:

    if "url" not in df.columns:
        print("A coluna 'url' não existe no DataFrame.")

    else:
        urls = df["url"].dropna()
        urls = urls[urls.astype(str).str.strip() != ""]

        if urls.empty:
            print("A coluna 'url' não possui URLs válidas.")

        else:
            for url in urls:
                registro = {
                    "request": str(url),
                    "status": "iniciado",
                    "status_code": None,
                    "response": None,
                    "erro": None
                }

                try:
                    response_files = requests.get(
                        str(url),
                        timeout=30
                    )

                    registro["status_code"] = response_files.status_code

                    response_files.raise_for_status()

                    files_data = response_files.json()
                    registro["response"] = files_data

                    if isinstance(files_data, list):
                        registro["status"] = "sucesso"
                    else:
                        registro["status"] = "resposta_inesperada"
                        registro["erro"] = (
                            "A resposta da API não é uma lista."
                        )

                except requests.exceptions.Timeout:
                    registro["status"] = "erro"
                    registro["erro"] = "Tempo limite da requisição excedido."

                except requests.exceptions.HTTPError as e:
                    registro["status"] = "erro_http"
                    registro["erro"] = str(e)

                except requests.exceptions.JSONDecodeError as e:
                    registro["status"] = "erro_json"
                    registro["erro"] = (
                        f"A resposta não contém JSON válido: {e}"
                    )

                except requests.exceptions.RequestException as e:
                    registro["status"] = "erro_requisicao"
                    registro["erro"] = str(e)

                except Exception as e:
                    registro["status"] = "erro_inesperado"
                    registro["erro"] = str(e)

                registros_log.append(registro)

else:
    print("Não há dados de contratação.")

log = json.dumps(
    registros_log,
    ensure_ascii=False,
    indent=2,
    default=str
)

print(log)
```

# Preparar para baixar arquivos

```python
from pathlib import Path
from urllib.parse import urlparse, unquote
import mimetypes
import re
import requests


def sanitizar_nome_arquivo(nome: str) -> str:
    """
    Remove caracteres inválidos do nome do arquivo.
    """
    nome = unquote(nome)
    nome = re.sub(r'[<>:"/\\|?*]', "_", nome)
    nome = nome.strip().strip(".")
    return nome or "arquivo_sem_nome"


def obter_nome_arquivo(
    response: requests.Response,
    url: str,
    indice: int
) -> str:
    """
    Tenta determinar o nome do arquivo usando:
    1. Content-Disposition;
    2. caminho da URL;
    3. Content-Type;
    4. nome genérico.
    """

    content_disposition = response.headers.get("Content-Disposition", "")

    # Exemplo:
    # attachment; filename="edital.pdf"
    # attachment; filename*=UTF-8''edital.pdf
    match = re.search(
        r"filename\*?=(?:UTF-8''|[\"']?)([^\"';]+)",
        content_disposition,
        flags=re.IGNORECASE
    )

    if match:
        return sanitizar_nome_arquivo(match.group(1))

    nome_url = Path(unquote(urlparse(url).path)).name

    if nome_url and "." in nome_url:
        return sanitizar_nome_arquivo(nome_url)

    content_type = response.headers.get(
        "Content-Type",
        ""
    ).split(";")[0].strip()

    extensao = mimetypes.guess_extension(content_type) or ""

    return f"arquivo_{indice:04d}{extensao}"


def gerar_caminho_unico(
    pasta: Path,
    nome_arquivo: str
) -> Path:
    """
    Evita sobrescrever arquivos com o mesmo nome.
    """

    caminho = pasta / nome_arquivo

    if not caminho.exists():
        return caminho

    stem = caminho.stem
    suffix = caminho.suffix
    contador = 1

    while True:
        novo_caminho = pasta / f"{stem}_{contador}{suffix}"

        if not novo_caminho.exists():
            return novo_caminho

        contador += 1


def baixar_arquivo(
    url: str,
    pasta_destino: Path,
    indice: int,
    session: requests.Session
) -> dict:
    """
    Baixa um único arquivo e retorna informações da operação.
    """

    resultado = {
        "transaction": "baixar-arquivos",
        "request": url,
        "status": "iniciado",
        "status_code": None,
        "arquivo": None,
        "erro": None
    }

    try:
        with session.get(
            url,
            stream=True,
            timeout=(10, 120)
        ) as response:

            resultado["status_code"] = response.status_code
            response.raise_for_status()

            nome_arquivo = obter_nome_arquivo(
                response=response,
                url=url,
                indice=indice
            )

            caminho_arquivo = gerar_caminho_unico(
                pasta=pasta_destino,
                nome_arquivo=nome_arquivo
            )

            with open(caminho_arquivo, "wb") as arquivo:
                for bloco in response.iter_content(
                    chunk_size=1024 * 1024
                ):
                    if bloco:
                        arquivo.write(bloco)

            resultado["status"] = "sucesso"
            resultado["arquivo"] = str(caminho_arquivo)

    except requests.exceptions.Timeout:
        resultado["status"] = "erro_timeout"
        resultado["erro"] = "Tempo limite excedido."

    except requests.exceptions.HTTPError as e:
        resultado["status"] = "erro_http"
        resultado["erro"] = str(e)

    except requests.exceptions.RequestException as e:
        resultado["status"] = "erro_requisicao"
        resultado["erro"] = str(e)

    except OSError as e:
        resultado["status"] = "erro_arquivo"
        resultado["erro"] = str(e)

    except Exception as e:
        resultado["status"] = "erro_inesperado"
        resultado["erro"] = str(e)

    return resultado
```

# Baixar arquivos

```python
pasta_destino = Path("arquivos_pncp")
pasta_destino.mkdir(parents=True, exist_ok=True)

resultados_download = []

with requests.Session() as session:
    session.headers.update({
        "User-Agent": "Mozilla/5.0 Python requests"
    })

    contador_arquivos = 1

    for registro in registros_log:

        if registro.get("status") != "sucesso":
            continue

        resposta_api = registro.get("response")

        if not isinstance(resposta_api, list):
            continue

        for item in resposta_api:

            if not isinstance(item, dict):
                continue

            url_arquivo = item.get("url")

            if not url_arquivo:
                continue

            resultado = baixar_arquivo(
                url=str(url_arquivo),
                pasta_destino=pasta_destino,
                indice=contador_arquivos,
                session=session
            )

            print(
                f"{resultado}"
            )

            resultados_download.append(resultado)

            contador_arquivos += 1
```

# Gravar arquivos de log

```python

with open(
    "log_downloads_pncp.json",
    "w",
    encoding="utf-8"
) as arquivo_log:

    json.dump(
        resultados_download,
        arquivo_log,
        ensure_ascii=False,
        indent=2
    )


log = json.dumps(
    resultados_download,
    ensure_ascii=False,
    indent=2,
    default=str
)

print(log)
```
