```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('aula_01_exemplo_01.csv', sep=",")
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['age'])
plt.title('Distribuição de Age')
plt.xlabel('Age')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='charges', data=df)
plt.title('Age vs Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', hue='smoker', y='charges', data=df)
plt.title('Age vs Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
correlation_matrix = df.corr(
    numeric_only=True,
    method='spearman'
)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix,
            annot=True,
            cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['bmi'])
plt.title('Distribuição do BMI')
plt.xlabel('BMI')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['children'], discrete=True)
plt.title('Distribuição children')
plt.xlabel('Children')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['charges'])
plt.title('Distribuição do charges')
plt.xlabel('charges')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(
        x='sex',
        y='charges',
        data=df,
        hue='sex',
        estimator=np.median,
        err_kws={'linewidth': 0}
)

plt.title('charges vs sex')
plt.xlabel('sex')
plt.ylabel('Mediana do charges')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.title('charges vs smoker')
plt.xlabel('smoker')
plt.ylabel('Mediana do charges')

sns.barplot(
    x='smoker',
    y='charges',
    data=df,
    hue='smoker',
    err_kws={'linewidth': 0},
    estimator=np.median
)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.barplot(
    x='region',
    y='charges',
    data=df,
    hue='region',
    err_kws={'linewidth': 0},
    estimator=np.median
)

plt.title('charges vs region')
plt.xlabel('region')
plt.ylabel('Mediana do charges')

plt.show()
```

```python
df['tem_filhos'] = (df['children'] > 0).astype(int)

df.head()
```

```python
plt.figure(figsize=(8, 6))

sns.barplot(
    x='tem_filhos',
    y='charges',
    data=df,
    hue='tem_filhos',
    err_kws={'linewidth': 0},
    estimator=np.median
)

plt.title('charges vs Tem Filhos')
plt.xlabel('Ten Filhos (0: Não, 1: Sim)')
plt.ylabel('Mediana do charges')

plt.show()
```
